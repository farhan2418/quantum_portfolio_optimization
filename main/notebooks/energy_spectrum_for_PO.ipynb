```python
# here we try to get the energy eigen values for the cost hamiltonian to get the energy gap between each connect node of hypercube graph
from qiskit.quantum_info import SparsePauliOp

cost_hamiltonian = SparsePauliOp(['IIIZ', 'IIZI', 'IZII', 'ZIII', 'IIZZ', 'IZIZ', 'ZIIZ', 'IZZI', 'ZIZI', 'ZZII'],
              coeffs=[-0.09513498+0.j,  0.27919293+0.j,  0.63435233+0.j, -0.08706848+0.j,
  1.65306991+0.j,  1.65396454+0.j,  1.65374103+0.j,  1.64886899+0.j,
  1.65352976+0.j,  1.65536693+0.j]) 

# since H_c is diagonal in the computational basis, the diagonal elements are the eigenvalues.
all_energies = cost_hamiltonian.to_matrix().diagonal().real
map1 = {}
for i, energy in enumerate(all_energies):
    binary_state = f"{i:0{cost_hamiltonian.num_qubits}b}"
    map1[energy] = binary_state
    print(f"State |{binary_state}> -> energy value: {energy:.4f}")
```
